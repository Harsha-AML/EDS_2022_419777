# Reading CSV File

```python
v_data=r"C:\Users\visha\Desktop\EDS\Harsha_Katikireddy_EDS\data\raw\time_series_covid19_vaccine_global.csv"
v_data= pd.read_csv(v_data)
```

```python
v_data.head()
```

# Deleting and Modifying unwanted columns and data

```python
data_1= v_data.drop(['Doses_admin', 'People_partially_vaccinated', 'Report_Date_String', 'UID', 'Province_State'], axis=1, inplace=True)
data_2=v_data.rename(columns={'Country_Region': 'Country', 'People_fully_vaccinated': 'Vaccinated_People'})
```

```python
data_2.head()
```

# Brazil's Vaccinated Persons per population

```python
dataframe_Brazil = data_2[data_2["Country"] == 'Brazil'].copy()
dataframe_Brazil['Vaccinated_People'] = dataframe_Brazil['Vaccinated_People'].div(21600000)
```

# Belgium's Vaccinated Persons per population

```python
dataframe_Belgium = data_2[data_2["Country"] == 'Belgium'].copy()
dataframe_Belgium['Vaccinated_People'] = dataframe_Belgium['Vaccinated_People'].div(11500000)
```

# Ireland's Vaccinated Persons per population

```python
dataframe_Ireland = data_2[data_2["Country"] == 'Ireland'].copy()
dataframe_Ireland['Vaccinated_People'] = dataframe_Ireland['Vaccinated_People'].div(5047288)
```

# Deleting the NaN values from the three countries

```python
Brazil = dataframe_Brazil.dropna()
Belgium = dataframe_Belgium.dropna()
Ireland = dataframe_Ireland.dropna()
```

# Creating a list of the three countries

```python
dataframe_Country = [Brazil, Belgium, Ireland]
```

```python
final_dataframe = pd.concat(dataframe_Country).reset_index(drop=True)
```

```python
final_dataframe
```

#Converting the date format
 

```python
final_dataframe['Date'] =  pd.to_datetime(final_dataframe['Date'], format='%Y-%m-%d')
```

```python
plot = pe.line(final_dataframe, x="Date", y="Vaccinated_People", color = 'Country', title='Vaccination Rate per population of Brazil, Belgium and, Ireland')
plot.show()
```
